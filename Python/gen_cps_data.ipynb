```python
### SETUP

import os, sys

import pandas as pd
pd.options.mode.chained_assignment = None

import warnings
warnings.simplefilter(action = 'ignore', category = pd.errors.PerformanceWarning)

code_dir = os.getcwd() + '/code/'
data_dir = os.getcwd() + '/data/'

sys.path.append(code_dir)

import cps_data as cd
import fred_data as fd
```

```python
### COLLECT CPS DATA (TIME AND MEMORY CONSUMING STEP)

cps_api_key = 'YOUR_CPS_API_KEY'

cps_df = cd.final_cps_df(cps_api_key, data_dir)
```

```python
### COLLECT STATE-SPECIFIC UR DATA

fred_api_key = 'YOUR_FRED_API_KEY'

sur_as_of_date = '3/20/2025'

sur_df = fd.all_fred_sur_df(fred_api_key, sur_as_of_date)
```

```python
### MERGE CPS AND 1-Y AVG STATE URs AND SAVE DATA

cps_df = cps_df.merge(sur_df[['yyyymm_state', 'sur_sa_1y_avg']], on = 'yyyymm_state', how = 'left')
cps_df.to_pickle(data_dir + '/cps_final.pkl')
```
